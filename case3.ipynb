```python
x = base_casa[['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number_of_Rooms',
       'Avg_Area_Number_of_Bedrooms', 'Area_Population', 'Price']]

y = base_casa[['Price']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x,y, train_size= 0.7,test_size= 0.3, random_state= 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r = r2_score(Y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
```

```python
print('r_quadrado',r)
```

```python
c = [i for i in range(1,1501,1)]
fig = plt.figure(figsize=(12,8))
plt.plot(c, Y_test, color ='red')
plt.plot(c, y_pred, color='blue')
plt.xlabel('index')
plt.ylabel('Preço do imovel');
```
